# **Bonds Pricing Components**

**Face Value or Par Value (FV):** The face is the **principal amount** of money loaned by the investor to the issuer. The amount is repaid to the investor at maturity.

**Coupons (C) or coupon rate (r):** Couponds are the interest received by an investor (bondholder) as they own a bond, paid by the issuer. The amount of interest paid is determeined by the coupon rate, and it is calculated on the face value of the bond. Coupon rates associated to bonds are scaled on current interest rates. When the central bank increases interest rate, newly issued bonds offer higher yield due to the new coupon rates associated. Roughly:

- If interest rates rises, coupon rate and yields rises whilst their bond prices falls.
- If interest rates drops, coupon and yields decreases whilst their bond prices increase. we compute coupon as:

$$C = FV * r.$$

**'r'** the coupon rate is an annual rate fixed at the bond issuance according to current interest rates. If it fluctuates the bond price will have to vary to meet sufficeint **yield** to attract investors.

**Yield to Maturity (YTM):** Yield to maturity is
calculated as an annual rate and represents the total expected return of the bond if held until maturity. It's the bonds **annualized return** and can be used as a metric to compare bonds with different coupon payments and maturities. Computed as follows:

$$YTM = \frac{C + \frac{FV - PV}{t} }{\frac{FV + PV}{2} }$$

Where:  
- **PV:** is the present value of the bond (sum of discounted cashflows, that is the discounted coupon payments plus the discounted face value).
$$PV = \sum_{t}^{T} \frac{C}{(1+r)^t}  + \frac{FV}{(1 + r)^t}$$
- **t:** ($t \in 1, 2,,...T$) the frequency or number of times per year interest is earned by the investor, it can be annual, seminannual, quarterly  etc,. Mostly ammually and semiannually.


**Dirty Price:** This is the price of a bond paid by the investor when buying the bond in the middle of two coupon payment dates. Since bond prices changes as coupon is paid across its duration, dirty price are destined to change. Thus the dirty price is the present value of the bond at time **t**.
$$DP = PV$$
With $DP$ = Dirty Price and, $PV$ = Present value of the bond.

**Accrued Interest (AI):** This is the accumullated Ã­nterest of the bond from its issuance date up to the purchase date. Formula:

$$AI = FV * r * \frac{t_{p} - t_{l} }{365} $$

With $AI$ = accrued interest. $FV$ = Face Value. r = coupon rate. $t_{p}$ = purchase date  $t_{l}$ = last accrual period. The formula is essentially the pro-rata date multiplied to coupons.

**Clean Price (CP):** This is the price of the bond that is quoted in the bond markets. It reflects the pure value of the bond bar past interests. This is computed by simply excluding accrued interest to the dirty price:

$$CP = DP - AI$$

with  $CP$ = Clean Price. $DP$ = Dirty Price. $AI$ = accrued interests. Clean price tends to vary over time with interest rates and issuuer's creditworthiness.



**Macaulay Duration (McD):**

Duration is used to assess a bond's sensitivity to **rate** changes (So is Macaulay Duration). By calculating the duration of the bond, we will later be able to compute the new price of a bond after a variation of its $YTM$. We compute it below (it is expressed in years):

$$McD = \frac{\sum_{t = 1}^{T} PV_{\tau} * \tau }{CP} $$

With $McD$ : Macaulay duration. $T$: (1, 2, 3,... T). $PV_{t}$: present value at time $t$.  $t$ : number of years between **payments date** and **purchase date**.  

**Modified Duration (MD):** to accurately calculate a bond price variation from a change in $YTM$, the $McD$ needs some adjustments.
$$MD = \frac{McD}{1 + YTM}$$.  

With a modified duration of 10, a bond price value would drop circa 10%, if the interest rate increases by 1%.

**Convexity (CVX):** A bonds convexity provides a measure of its duration change as interest rates fluctuate. It could be seen as the 'curvature' of the relationship between a bond price and its yield. The higher the convexity the lower the the sensitivity of a bond to interest rates changes. Thus higher convex bonds offer a better protection against shifting interest rates compared to lower convexity ones. Formula:

$$CVX = \frac{\sum_{t = 1}^{T} PV_{t}  * t^2}{CP}$$

$\tau^2$ : squared number of years between **payments date** and **purchase date**.  

Convexity is a great complementary tool to calculate bond price changes. Indeed, when interet rates changes, the prices of a bond does not vary linearly. Calculating the convexity allows us to take into account the variation of a bonds duration as interest's rates fluctuates.

**Bond Price Variation:** Using duration and convexity, we can now calculate the percentage variation that will occur on a bond price if we observe a change in interest rates. Formula:

$$\Delta CP = (- MD * \Delta{YTM} ) + 0,5 * CVX * \left\{\frac{\Delta{YTM} }{100} \right\}^2 *  100 $$

with $\Delta{CP}$: bond price variation in %. $MD$ : Modified duration. $\Delta{YTM}$: Yield to maturity variation (interest rates change) in %. $CVX$ = Convexity.


After getting this value: Bond Price Variation, we can compute the bond's clean price ($DP$ - $AI$) Present value minus Accrued interest) for the newly computed change in its yield to maturity. By just applying this percentage to calculate the new bond price:
$$CP' = CP * ( 1 + \Delta{CP})$$

where: $CP'$ = New bond price. $CP$: bond clean price. $\Delta{CP}: bond price variation in %$

The above formulae are essential for understanding the bonds primary and secondary markets. It also depitcs the relationship between bond's prices and interest rates changes by central banks.

```python
# define bond price function
def bond_pricing(face_value, coupon_rate, freq, yield_to_maturity, purchase_date, next_coupon_date, maturity_date):

  """


 Yield to Maturity (YTM): ,... should be inputed based on the freq.
  etc
    ....

  """



  # convert datess
  purchase_date =   datetime.strptime(purchase_date, "%Y-%m-%d")
  next_coupon_date = datetime.strptime(next_coupon_date, "%Y-%m-%d")
  maturity_date = datetime.strptime(maturity_date, "%Y-%m-%d")


  starting_date = next_coupon_date
  date_rows = [next_coupon_date]   # Initialize new date list with 'next coupon date' as first date


  ## Obtain incremental dates (based on frequency) from next coupon payment date until maturity of the bond.
  while starting_date < maturity_date:
    if freq == 1:
      starting_date = starting_date.replace(year = starting_date.year + 1)
      date_rows.append(starting_date)
    elif freq == 2:
      starting_date = starting_date + relativedelta(months = 6)
      date_rows.append(starting_date)
    else:
      raise ValueError("Compunding Frequency must be either (annual) 1 or (semiannual) 2\
       or just go and write a new function to accomadate other compunding frequencies")


  ## We obtain time to maturity - t
  t_list = []
  for date in date_rows:
    t_list.append((date - purchase_date).days / 365)

  ## Convert date back to string, so to print more legible dates
  date_rows = [dt.strftime("%Y-%m-%d") for dt in date_rows]

  ## We compute coupons (C) and present values (PV)
  coupons_rows = []
  for period in range(len(date_rows)):
    coupons_rows.append(face_value * coupon_rate)
  coupons_rows[-1] += face_value         # Last coupon is received plus face value of the bond

  pv_rows = []
  for period, coupon in zip(t_list, coupons_rows):
    pv_rows.append(np.round(coupon / (1 + yield_to_maturity) ** period, 2) ) # discounting cash flows

  ## Payment dataframe
  payments_data = [date_rows, coupons_rows, pv_rows]
  payment_columns = ['Payment Dates', 'Coupons', 'Present Values']
  payment_df = pd.DataFrame(payments_data, payment_columns)
  payment_df = payment_df.transpose()

  #### Compute Accrued interest, Duration and convexity
  # Present value mulytiplied by times: PV * t
  pv_times_t = []
  for pv, period in zip(pv_rows, t_list):
    pv_times_t.append(pv * period)

  # Present value mulytiplied by times squared: PV * t^2
  tsquared = [t ** 2 for t in t_list]
  pv_times_tsquared = []
  for pv, t_squared in zip(pv_rows, tsquared):
    pv_times_tsquared.append(pv * t_squared)

  # Aggregate data
  calculations_data = [date_rows, pv_times_t, pv_times_tsquared]
  calculations_columns = ['Payment Dates', 'PV * t', 'PV * t^2']
  calculations_df = pd.DataFrame(calculations_data, calculations_columns)
  calculations_df = calculations_df.transpose()

  # Accrued interest
  if freq == 1:
    last_coupon_date = next_coupon_date.replace(year = next_coupon_date.year - 1)
  elif freq == 2:
    last_coupon_date = next_coupon_date.replace(month = next_coupon_date.month - 6)

  accrued_interest = np.round((face_value * coupon_rate) * (purchase_date - \
                                                  last_coupon_date).days / 365 )
  dirty_price = np.round( payment_df['Present Values'].sum(), 2 )
  clean_price = dirty_price - accrued_interest

  mc_duration = np.round( (payment_df['Present Values'] * t_list).sum() / clean_price, 2)
  md = np.round(mc_duration / (1 + yield_to_maturity), 2)
  cvx = np.round( (payment_df['Present Values'] * tsquared).sum() / clean_price, 2)

  bond_pricing_data = [dirty_price, accrued_interest, clean_price, mc_duration, md, cvx]
  bond_pricing_columns = ['Dirty Price', 'Accrued Interest', 'Clean Price', 'Macaulay Duration', 'Modified Duration', 'Convexity']
  bond_pricing_df = pd.DataFrame(bond_pricing_data, bond_pricing_columns)
  bond_pricing_df = bond_pricing_df.transpose()

  return payment_df, bond_pricing_df
```

```python
# Define bond price fluctuation function.
def bond_price_fluctuation(change_in_ytm):
  """
  """

  # Extract values from bond pricing df
  modified_duration = pricing_df['Modified Duration'][0]
  convexity = pricing_df['Convexity'][0]
  clean_price = pricing_df['Clean Price'][0]

  # Compute new bond price
  change_in_bond_price = np.round( (- modified_duration * change_in_ytm) + 0.5 \
                        * convexity * (change_in_ytm /100 ) ** 2 * 100, 4 )
  new_bond_price = np.round( clean_price * (1 + change_in_bond_price), 2)


  fluctuation_data = [change_in_ytm, change_in_bond_price, new_bond_price]
  fluctuation_columns = ['Change in YTM', 'Change in Bond Price', 'New Bond Price']

  bond_price_fluctuation_df = pd.DataFrame(fluctuation_data, fluctuation_columns)
  bond_price_fluctuation_df = bond_price_fluctuation_df.transpose()

  return bond_price_fluctuation_df
```
